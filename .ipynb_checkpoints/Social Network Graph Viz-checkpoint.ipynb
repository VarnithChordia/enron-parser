```python
import json

with open('json.json') as data_file:    
    data = json.load(data_file)

data
```

```python
senders = {}
for entry in data:
    if entry["sender"] in senders:
        senders[entry["sender"]] += 1
    else:
        senders[entry["sender"]] = 1
senders
```

```python
emails = []

for entry in data:
    if entry["sender"] not in emails:
        emails.append(entry["sender"])
    for r in entry["recipient"]:
        if r not in emails:
            emails.append(r)
        
email_to_pos_map = {}
for i, email in enumerate(emails):
    email_to_pos_map[email] = i

email_to_pos_map
```

```python
import numpy as np

network_graph = np.zeros((len(emails), len(emails)))

for entry in data:
    for r in entry["recipient"]:
        network_graph[email_to_pos_map[entry["sender"]], email_to_pos_map[r]] += 1
    
network_graph
```

```python
import networkx as nx
import matplotlib.pyplot as plt 
nx.draw(nx.from_numpy_matrix(network_graph))


plt.show()
```
